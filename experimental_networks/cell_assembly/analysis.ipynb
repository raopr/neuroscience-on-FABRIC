```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 8})
```

```python
within = pd.read_csv("within.csv", index_col = 0)
fig, ax = plt.subplots(2, 2, figsize = (10, 10))
counter = 0
for i in range(2):
    for j in range(2):
        ax[i, j].matshow(within.loc[f"{i}-exc", :].to_numpy().reshape(5, 5))
        ax[i, j].set_title(f"assembly {counter}")
        ax[i, j].set_xlabel("cell local id")
        ax[i, j].set_ylabel("cell local id")
        counter += 1

fig.suptitle("exc")
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10, 10))
counter = 0
for i in range(2):
    for j in range(2):
        ax[i, j].matshow(within.loc[f"{i}-inh", :].to_numpy().reshape(5, 5))
        ax[i, j].set_title(f"assembly {counter}")
        ax[i, j].set_xlabel("cell local id")
        ax[i, j].set_ylabel("cell local id")
        counter += 1

fig.suptitle("inh")
```

```python
between = pd.read_csv("between.csv", index_col = 0)
print(between.shape)

fig, ax = plt.subplots(1, 2, figsize = (10, 20))
for i, name in enumerate(["exc", "inh"]):
    ax[i].matshow(between.loc[name, :].to_numpy().reshape(20, 20))
    ax[i].set_title(name)
    ax[i].set_xlabel("cell gid")
    ax[i].set_ylabel("cell gid")
```
