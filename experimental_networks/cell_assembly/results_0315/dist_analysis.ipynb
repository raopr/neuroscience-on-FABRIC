```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random = np.array([647.31, 523.52, 262.64, 58.98])
assemb = [649.55, 524.24, 262.19, 58.99]
cols = ["CPU (1N)", "1 GPU (1N)", "2 GPU (1N)", "4 GPU (2N)"]
ngpus = np.array([0, 1, 2, 4])
```

```python
plt.plot(ngpus, random)
plt.scatter(ngpus, random)
plt.xticks(ticks = ngpus, labels = cols)
plt.grid()
plt.ylabel("Seconds")
```

```python
# For small networks
1. Increasing the number of GPUs gives a faster-than-linear decrease in sim time.
2. By-assembly and random distribution give approximately the same results.
3. 2 GPUs (1N) = 2 GPUs(2N), so the bridge is very fast.
4. The bridge between sites is about 23% slower (2GPU 2S vs 2 GPU 1N)
```
